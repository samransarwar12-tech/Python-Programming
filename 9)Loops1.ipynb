# While loop

```python
x = 1
while x < 10:
  print(x)
  x += 1
```

```python
x = 1
while x < 10:
  print(x)
  x += 2
```

```python
x = 1
while x < 10:
  print(x)
  x += 3
```

```python
x = 1
while x < 10:
 print(x)
 if x==5:
  break
 x += 1
```

```python
x = 1
while x < 10:
  x += 1
  if x == 5:
    continue
  print(x)
```

```python
x = 1
while x < 10:
  x += 1
  print(x)
  if x == 5:
    continue
```

```python
x = 1
while x in range(10):
  print(x)
  x += 1
```

```python
x = 1
while x < 5 :
    y = 1
    while  y < 5 :
        print('x->',x,'y->',y,'x + y ->',x+y)
        y += 1
    print("\n")
    x += 1
```
