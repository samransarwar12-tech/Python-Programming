```python
name = input("Enter your full name: ")
age = int(input("Enter your age: "))
course = input("Enter your course name: ")
fees = float(input("Enter course fees: "))

first_letter = name[0]
last_letter = name[-1]

discount = fees * 0.10
final_fees = fees - discount

print("\n----- STUDENT REPORT -----")

print(f"Name : {name.title()}")
print(f"Age : {age}")
print(f"Course : {course.upper()}")

print(f"\nFirst Letter : {first_letter}")
print(f"Last Letter : {last_letter}")

print(f"\nOriginal Fees : ₹{fees}")
print(f"Discount : ₹{discount}")
print(f"Final Fees : ₹{final_fees}")

print("\n----- DATA TYPES -----")
print(type(name))
print(type(age))
print(type(fees))

print(f"\nWelcome {name.title()}")
```
