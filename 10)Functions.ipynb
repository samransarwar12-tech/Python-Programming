```python
def name1(lname,fname):
  print(fname + " " + lname)

name1(fname = 'samran',lname = 'ali')
```

```python
def name2(**student):
  print(student['fname'] + " " + student['lname'])

name2(fname = 'samran',lname = 'ali')
```

```python
def name3(**students):
  print(students['s1'],students['s2'],students['s3'])

name3(s1 = 'samran',s2 = 'ali',s3 = 'abdullah')
```
